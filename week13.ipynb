```python
# Import necessary libraries
from tensorflow.keras.applications import VGG16
from tensorflow.keras.utils import plot_model

# Step 1: Load the pretrained model
model = VGG16(weights='imagenet')

# Step 2: Display model summary
model.summary()

# Step 3: Print total parameters
total_params = model.count_params()
print("\nTotal Parameters:", total_params)

# Step 4: Inspect layers
print("\nLayer Details:")
for i, layer in enumerate(model.layers):
    print(f"{i+1}. Layer Name: {layer.name}, Type: {layer.__class__.__name__}, Trainable: {layer.trainable}")

# Step 5: Visualize the architecture (optional)
plot_model(model, to_file='vgg16_architecture.png', show_shapes=True, show_layer_names=True)
```
