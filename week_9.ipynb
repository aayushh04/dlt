```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Parameters
vocab_size = 1000      # Size of the vocabulary
max_length = 100       # Maximum length of input sequences
embedding_dim = 64     # Dimension of embedding vectors
rnn_units = 128        # Number of RNN units
num_classes = 2        # For binary classification

# Dummy data: 1000 sequences of integers
X = np.random.randint(0, vocab_size, size=(1000, max_length))
y = np.random.randint(0, num_classes, size=(1000,))

# Build the RNN model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    SimpleRNN(rnn_units),
    Dense(num_classes, activation='softmax')  # Use 'sigmoid' for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=5, batch_size=32)

# Evaluate
loss, acc = model.evaluate(X, y)
print(f"Training Accuracy: {acc:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated training loss values over 20 epochs
epochs = 20
np.random.seed(42)
loss_values = np.exp(-0.3 * np.arange(epochs)) + 0.05 * np.random.randn(epochs)
loss_values = np.clip(loss_values, a_min=0, a_max=None)  # Loss cannot be negative

# Plotting the training loss curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs + 1), loss_values, marker='o', color='blue')
plt.title('Simulated RNN Training Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Parameters (same as before)
vocab_size = 1000
max_length = 100
embedding_dim = 64
rnn_units = 128
num_classes = 2

# Create dummy training data
X_train = np.random.randint(0, vocab_size, size=(1000, max_length))
y_train = np.random.randint(0, num_classes, size=(1000,))

# Create dummy test data
X_test = np.random.randint(0, vocab_size, size=(10, max_length))  # 10 samples

# Build the model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    SimpleRNN(rnn_units),
    Dense(num_classes, activation='softmax')
])

# Compile and train
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=32)

# Predict on test data
predictions = model.predict(X_test)

# Show predicted classes for each test sample
predicted_classes = np.argmax(predictions, axis=1)
print("Predicted classes:", predicted_classes)
```
