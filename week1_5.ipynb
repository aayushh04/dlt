```python
# ======================================
# Lab 15: YOLOv8 detection (Ultralytics)
# Train/Val/Predict on built-in COCO8
# ======================================
!pip -q install ultralytics

from ultralytics import YOLO
import torch, os, glob
from IPython.display import Image, display

torch.manual_seed(42)
print("Torch CUDA:", torch.cuda.is_available())

# Load nano model for speed (transfer learning)
model = YOLO('yolov8n.pt')

# Train on tutorial dataset (quick demo) â€” increase epochs for better mAP
results = model.train(data='coco8.yaml', epochs=5, imgsz=640, device=0 if torch.cuda.is_available() else 'cpu')

# Validate
val = model.val()

# Inference on a sample image
sample_candidates = glob.glob('/content/ultralytics/coco8/images/val/*.jpg')
sample_img = sample_candidates[0] if sample_candidates else None
if sample_img:
    preds = model.predict(source=sample_img, conf=0.25)
    out_dir = preds[0].save_dir
    for p in glob.glob(os.path.join(out_dir, '*.jpg')):
        display(Image(filename=p))
else:
    print("Sample image not found.")
```
